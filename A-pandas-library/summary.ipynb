import pandas as pd

#preparando los datos y los nombres de las columnas
atlas = [
      ['France', 'Paris'],
        ['Russia', 'Moscow'],
        ['China', 'Beijing'],
        ['México', 'Ciudad de México'],  
        ['Egypt', 'Cairo'],
]
geography = ['country', 'capital']

#creando un DataFrame
world_map = pd.DataFrame(data=atlas , columns=geography)

print(world_map.dtypes) #imprimimos el atributo dtypes
print(world_map.columns) #imprimimos el atributo columns 
print(world_map.shape) #imprimimos el atributo shape 
print(world_map.info()) #imprimimos todos los atributos a la vez

¿Cómo es la notación abreviada para la indexación desde la variable result?

result = music_df.loc[5:]

*result = music_df[5:]
Este slice contendrá filas a partir del índice 5. No hay un atributo loc en la notación abreviada y, en este caso, el slice seguirá siendo el mismo. Pero esto no siempre es así. En la notación abreviada, el final del rango no se incluye en el slice. Por ejemplo, la línea con índice 5 no se incluirá en el slice music_df[:5].

result = music_df[:5]

result = music_df[:, 5:]
