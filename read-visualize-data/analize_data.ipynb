¿Cómo puedo comenzar a analizar mis datos?

El primer paso para comenzar a analizar tus datos es obtener una vista preliminar de los mismos. Con una visión general de éstos, podemos entender su estructura general, identificar problemas potenciales y planificar tu análisis, por mencionar algunos de sus beneficios.

Lo que aprenderás
Conocer y aplicar diferentes métodos para obtener una visón general de un dataset.

El métodoinfo()

El método info()es una de las funcionalidades más básicas en un Dataframe. Esta función nos entrega información relacionada con las columnas que comprenden nuestro dataset, tales como:

El número total de filas y columnas.
El nombre de cada columna.
El número de entradas no nulas por columna.
El tipo de datos de cada columna (por ejemplo: int64, float64, object).
La cantidad de memoria utilizada por el DataFrame.
Veamos el siguiente ejemplo:

import pandas as pd

column_names = [
    'country',
    'name',
    'capacity_mw',
    'latitude',
    'longitude',
    'primary_fuel',
    'owner'
]
data = pd.read_csv(
    '/datasets/gpp_modified.csv',
    sep='|',
    header=None,
    names=column_names,
    decimal=',',
)

data.info()

Observa que el resultado del método info() no devuelve ningún valor con el que puedas interactuar directamente. Este método solo muestra un resumen de información sobre el DataFrame en la consola. Sin embargo, si deseas almacenar información similar a la que proporciona info(), puedes hacerlo accediendo a los atributos del DataFrame directamente, por ejemplo: dtypes, shape o isnull().

El método sample()

Mediante el método sample() podemos obtener una muestra aleatoria de las filas de nuestro DataFrame. Con esta funcionalidad, podemos obtener una vista representativa de nuestros datos así como una noción inicial de la distribución de los mismos.

import pandas as pd

column_names = [
    'country',
    'name',
    'capacity_mw',
    'latitude',
    'longitude',
    'primary_fuel',
    'owner'
]
data = pd.read_csv(
    '/datasets/gpp_modified.csv',
    sep='|',
    header=None,
    names=column_names,
    decimal=',',
)

print(data.sample(5))

Resumen
En esta lección, aprendimos a realizar un reconocimiento preliminar de nuestros datos utilizando métodos como info(), sample(), y head(). Estas funciones nos permiten entender la estructura del dataset, verificar la cantidad de registros, identificar tipos de datos y detectar posibles valores faltantes o inconsistencias. 

Conocer estos conceptos es crucial para comenzar cualquier análisis de datos, ya que garantiza que trabajemos con datos precisos y estructurados, lo que mejora la calidad y la confiabilidad de lo que busquemos desarrollar de aquí en adelante.

Terminología
A continuación, se resumen algunos de los atributos y métodos vistos, juntos con otros que también te serán de utilidad.

Método/Atributo	Descripción	Salida

info()	Proporciona un resumen conciso del DataFrame. Incluye número de filas, columnas, tipos de datos y valores no nulos.	Muestra información en consola.

head(n)	Muestra las primeras n filas del DataFrame (por defecto, las primeras 5 filas).	DataFrame con las primeras n filas.

sample(n)	Muestra una muestra aleatoria de n filas del DataFrame.	DataFrame con n filas seleccionadas al azar.

shape	Devuelve una tupla con el número de filas y columnas del DataFrame.	Tupla (num_filas, num_columnas).

dtypes	Muestra los tipos de datos de cada columna.	Serie con nombres de columnas y tipos de datos.

columns	Devuelve los nombres de las columnas del DataFrame.	Índice con nombres de las columnas.

isnull().sum()	Cuenta el número de valores nulos en cada columna.	
Serie con nombres de columnas y conteo de valores nulos.

notnull().sum()	Cuenta el número de valores no nulos en cada columna.	
Serie con nombres de columnas y conteo de valores no nulos.
