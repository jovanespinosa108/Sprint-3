Ejercicio 1

Ahora trabajaremos con un DataFrame simulado que contiene información sobre las ventas en diferentes sucursales de una empresa. Nuestro objetivo es utilizar la indexación lógica avanzada con loc para filtrar los datos y obtener información específica del negocio.

DataFrame:

sucursal	producto	cliente	cantidad	precio_unitario	venta_total
Centro	Camiseta roja	Ana García	2	15.99	31.98
Norte	Zapatillas azules	Luis Rodríguez	1	49.99	49.99
Sur	Camisa blanca	Pedro Hernández	3	29.99	89.97
Centro	Jeans negro	María López	1	39.99	39.99
Norte	Camiseta roja	Juan Pérez	4	15.99	63.96

Utilizando el método loc, filtre el DataFrame para obtener las ventas que cumplan con las siguientes condiciones en este orden:

Sucursal: sucursales "Centro" y "Norte" únicamente.
Producto: solo productos vendidos a un precio unitario mayor que 25.

```python
import pandas as pd

data = {
    'sucursal': ['Centro', 'Norte', 'Sur', 'Centro', 'Norte'],
    'producto': ['Camiseta roja', 'Zapatillas azul', 'Camisa blanca', 'Jeans negro', 'Camiseta roja'],
    'cliente': ['Ana Garcia', 'Luis Rodriguez', 'Pedro Hernandez', 'Maria Lopez', 'Juan Perez'],
    'cantidad': [2, 1, 3, 1, 4],
    'precio_unitario': [15.99, 49.99, 29.99, 39.99, 15.99],
    'venta_total': [31.98, 49.99, 89.97, 39.99, 63.96]
}

df = pd.DataFrame(data)

ventas_filtradas_df = df.loc[(df['sucursal'] == 'Centro') | (df['sucursal'] == 'Norte')]
ventas_filtradas_df = ventas_filtradas_df[ventas_filtradas_df['precio_unitario'] > 25]

print(ventas_filtradas_df)
```

Indexación lógica (booleana)

La indexación puede ser más eficaz si se incluye una expresión lógica. 
En el capítulo "Sentencias condicionales" escribiste algunos filtros sin utilizar pandas. 
Este es el procedimiento que seguiste:

hiciste un bucle a partir de una lista anidada;
utilizaste if para comprobar si se cumplió la condición;
realizaste la acción deseada.

En pandas, este complicado algoritmo puede llevarse a cabo en una línea de código. 
El procedimiento se denomina indexación lógica o indexación booleana y funciona de la siguiente manera:

Primero, encontraremos todas las filas en las que se cumple cierta condición. Por ejemplo, 
podríamos comprobar si la columna 'genre' es igual a 'pop' basándonos en el conjunto de 
datos music_log_chpt_11.csv con el que hemos trabajado anteriormente. 
Podemos utilizar la notación abreviada de esta forma:

 print(df['genre'] == 'pop')
 
 Otra alternativa es utilizar el atributo loc:

 print(df.loc[:, 'genre'] == 'pop')

 Como puedes ver, para cada fila (o índice) en la columna 'genre', obtenemos un valor booleano. 
 True indica que la condición se cumplió en esa fila en particular, lo que significa que 'genre' es en realidad 'pop'.

Ahora podemos utilizar el resultado obtenido en el paso anterior para filtrar la tabla original. 
Debemos obtener una tabla filtrada que contenga solamente canciones 'pop'. Para hacerlo,

 También podemos utilizar la notación abreviada, eliminando el método loc. De esta forma:

 print(df[df['genre'] == 'pop'])

Esta forma de indexación lógica es muy común. Puedes utilizar expresiones lógicas con cualquier operación lógica:

Para comprobar la igualdad o la desigualdad, respectivamente, utiliza == y !=.

Para comprobar si un valor es mayor o menor, utiliza > o <, respectivamente.

Utiliza >= para comprobar si un valor es mayor o igual que y <= para comprobar si un valor es menor o igual que.

Todas las expresiones lógicas devolverán los valores lógicos True o False.

Por cierto, para aplicar varias condiciones, primero podemos aplicar la primera condición y almacenar el resultado. 
Luego, podemos llamar a la indexación lógica con la segunda condición.
