Introducción

Esta es una recapitulación rápida de lo que has aprendido:

Exploraste cómo funciona la indexación con Pandas y aplicaste técnicas de indexación basadas tanto en etiquetas como en enteros:

import pandas as pd

# Acceder a los datos mediante sus etiquetas con loc
row_label = df.loc[5]

# Acceder a los datos mediante su posición con iloc
row_position = df.iloc[0]

También dominaste el arte de filtrar datos utilizando el método query() para condiciones más complejas:

# Filtrar filas donde 'age' es más grande que 30 utilizando query
filtered_data = df.query('age > 30')

Además, aprendiste a utilizar el método where() para reemplazar valores con base en condiciones:

import numpy as np

# Reemplazar valores en los que 'age' es menor que 0 con NaN
cleaned_data = df.where(df['age'] >= 0, np.nan)

Estas técnicas potenciarán tu capacidad para manipular y analizar tus datos de forma más efectiva. ¡Ahora, prueba tu comprensión con los cuestionarios!

¿Cuántos índices hay por defecto para un objeto Series y para un objeto DataFrame?

*Uno y uno.
Los objetos Series y DataFrame en pandas siempre tienen índices, que se almacenan en el atributo index (etiquetas de eje para Series y etiquetas de fila para DataFrame). Los DataFrames tienen técnicamente dos índices, pero cuando se utiliza el atributo index solo se hace referencia a las etiquetas de las filas. Cada vez que creas un Series o un DataFrame, su atributo de index se crea automáticamente con valores por defecto si no se especifican los valores del índice.

El índice es opcional para ambos, así que cero.

Elige tantas como quieras

No es posible sin fusionar ambos DataFrames en uno solo.

Construir una lista de valores de city_id con poblaciones superiores a 100 000, hacer una consulta de clients con city_id en esa lista.
large_cities = cities['population'] > 100000
clients_in_large_cities = clients[clients['city_id'].isin(large_cities)]

*Construir una lista de valores de city_id con poblaciones superiores a 100 000, hacer una consulta de clients con city_id en esa lista.

clients_in_large_cities = clients[clients['city_id'].isin(large_cities)]
Podemos filtrar el DataFrame con la lista de valores creada a partir de otro.
